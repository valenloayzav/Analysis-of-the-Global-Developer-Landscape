```python
import matplotlib.pyplot as plt
import seaborn as sns

try:
    current_languages = df['LanguageHaveWorkedWith'].dropna()
    current_counts = current_languages.str.split(';').explode().value_counts()
    top_10_current = current_counts.head(10)

    plt.figure(figsize=(12, 8))
    sns.barplot(x=top_10_current.values, y=top_10_current.index, palette='viridis')
    plt.title('Top 10 Most Used Programming Languages (Current Year)', fontsize=16)
    plt.xlabel('Number of Respondents', fontsize=12)
    plt.ylabel('Programming Language', fontsize=12)
    for index, value in enumerate(top_10_current.values):
        plt.text(value, index, f' {value}', va='center')
    plt.tight_layout()
except Exception as e:
    print(f"An error occurred while creating the current languages chart: {e}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

try:
    dev_roles = df['DevType'].dropna().str.split(';').explode()

    top_10_roles = dev_roles.value_counts().head(10)

    # Create the plot
    plt.figure(figsize=(12, 8))
    sns.barplot(x=top_10_roles.values, y=top_10_roles.index, palette='cubehelix')
    plt.title('Top 10 Most Common Developer Roles', fontsize=16)
    plt.xlabel('Number of Respondents', fontsize=12)
    plt.ylabel('Developer Role', fontsize=12)

    for index, value in enumerate(top_10_roles.values):
        plt.text(value, index, f' {value}', va='center')

    plt.tight_layout()
    plt.savefig('top_10_developer_roles.png')

    print("Chart for developer roles saved as 'top_10_developer_roles.png'")

    roles_df = top_10_roles.reset_index()
    roles_df.columns = ['Developer Role', 'Count']
    print("\nTop 10 Developer Roles Data:")
    print(roles_df)

except KeyError:
    print("Error: 'DevType' column not found in the dataset.")
except Exception as e:
    print(f"An error occurred while creating the developer roles chart: {e}")
```

```python
file_name = "/Users/valentinaloayza/Downloads/job-postings.csv"
df = pd.read_csv(file_name)
print("CSV file loaded successfully.")
```

```python
df_sorted = df.sort_values(by="Number of Job Postings", ascending=False)

plt.figure(figsize=(10,6))
bars = plt.bar(df_sorted["Location"], df_sorted["Number of Job Postings"], color="blue", edgecolor="black")

plt.title("Job Postings by Location", fontsize=16, fontweight="bold")
plt.xlabel("Location", fontsize=12)
plt.ylabel("Number of Job Postings", fontsize=12)
plt.xticks(rotation=45, ha="right")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 50,
             f"{int(height):,}",
             ha='center', va='bottom', fontsize=10, fontweight="bold", color="black")

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_name = "/Users/valentinaloayza/Downloads/popular-languages.csv"
df = pd.read_csv(file_name)
print("CSV file loaded successfully.")
```

```python
df["Annual Average Salary"] = df["Annual Average Salary"].replace('[\$,]', '', regex=True).astype(float)

df_sorted = df.sort_values(by="Annual Average Salary", ascending=True)

plt.figure(figsize=(10,6))
bars = plt.barh(df_sorted["Language"], df_sorted["Annual Average Salary"],
                color="blue", edgecolor="black")

plt.title("Annual Average Salary by Programming Language", fontsize=16, fontweight="bold")
plt.xlabel("Annual Average Salary (USD)", fontsize=12)
plt.ylabel("Programming Language", fontsize=12)

for bar in bars:
    width = bar.get_width()
    plt.text(width + 2000,
             bar.get_y() + bar.get_height()/2,
             f"${width:,.0f}",
             va='center', fontsize=10, fontweight="bold")


plt.tight_layout()

plt.show()
```
